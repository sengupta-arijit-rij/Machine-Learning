```python
# Imports
import pandas as pd
```

```python
#Load DataSet
data_url='https://raw.githubusercontent.com/pixeltests/datasets/main/aimes_housing.csv'
House=pd.read_csv(data_url)
```

```python
# View DataSet
House.head()
House.info()
```

Task 2:

```python
# Creating a subset copy
bath_Subset=House[["BsmtFullBath","BsmtHalfBath","FullBath","HalfBath"]].copy()
bath_Subset.head()
```

```python
# Creating total no of bathrooms
bath_Subset['Total_Bathrooms']= bath_Subset['BsmtFullBath'] + bath_Subset['BsmtHalfBath'] +bath_Subset['FullBath'] + bath_Subset['HalfBath']
bath_Subset.head()
```

Task 3:

```python
# Creating new subset for LotArea
lot_Subset=House[["LotArea"]]
lot_Subset.head()
```

```python
# Creating LargeLot
lot_Subset.loc[lot_Subset['LotArea']>=10000, 'Large_Lot']=1
lot_Subset.loc[lot_Subset['LotArea']<10000, 'Large_Lot']=0
lot_Subset.head()
```

Task 4:

```python
# Creating new subset for MoSold
MoSold_Subset= House[['MoSold']]
MoSold_Subset.head()
```

```python
# Creating Summer_Sale
# MoSold_Subset.loc[((MoSold_Subset['MoSold']>=1) &(MoSold_Subset['MoSold']<=5)) | ((MoSold_Subset['MoSold']>=9) &(MoSold_Subset['MoSold']<=12)),'Summer_Sale']=0
# MoSold_Subset.loc[((MoSold_Subset['MoSold']==6)|(MoSold_Subset['MoSold']==7)|(MoSold_Subset['MoSold']==8)),'Summer_Sale']=1

MoSold_Subset['Summer_Sale']= MoSold_Subset['MoSold'].apply(lambda x:0 if((1<=x<=5)|(9<=x<=12)) else 1)
MoSold_Subset.head(20)
```

```python
# Create function to accept a string in yyyy-mm format and return the year
def getYear(date):
  fields_date=date.split('-')
  month_mm=fields_date[0]  
  year_yyyy=fields_date[1]  
  return year_yyyy

random_Date='02-1987'
print('Year of reandom date '+random_Date+' is '+getYear(random_Date))
```

Task 8:

```python
# Apply created function to earliest_cr_line column
loans['earliest_cr_line']
loans=loans.copy()
loans['earliest_cr_line'].head()
```

```python
# calling the function for all values in column earliest_cr_line and storing in Earliest_Credit_Year
loans['Earliest_Credit_Year']=loans['earliest_cr_line'].apply(getYear)
loans[['Earliest_Credit_Year','earliest_cr_line']].head()
```
