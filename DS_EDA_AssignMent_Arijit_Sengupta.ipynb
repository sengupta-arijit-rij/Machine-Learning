```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load datasheet
data_url='https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
clm_headers=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week','native-country', 'income']
Adults= pd.read_csv(data_url, names=clm_headers)
```

```python
# View Dataset
Adults.head()
```

```python
# Assigning head and tail records
Adults_head= Adults.head(10)
Adults_Tail= Adults.tail(10)
```

```python
# Datatypes on Adults
Adults.dtypes
Adults.describe(include='number')
Adults.describe(include='object')
```

```python
# Missing values check
Adults.isnull().sum()
```

```python
# Describe mean age
print(Adults['age'].describe(include='number'))
mean_age=(Adults['age'].describe(include='number')).get(key='mean')
mean_age
```

```python
# Describe hours per week
print(Adults['hours-per-week'].describe(include='number'))
std_hpw=(Adults['hours-per-week'].describe(include='number')).get(key='std')
std_hpw
```

```python
# Describing non numericals
Adults.describe(exclude='number')
print(Adults['education'].describe(exclude='number'))
print('*'*20)
print(Adults['income'].describe(exclude='number'))
print('*'*20)
unique_edu=(Adults['education'].describe(exclude='number')).get(key='unique')
print(unique_edu)
print('*'*20)
freq_income=(Adults['income'].describe(exclude='number')).get(key='freq')
freq_income
```

```python
# Analyszing relationship column
print(Adults['relationship'].describe(exclude='number'))
print('*'*20)
print((Adults['relationship'].describe(exclude='number')).unique())
print('*'*20)
print((Adults['relationship'].describe(exclude='number')).value_counts())
```

```python
# Other relationship count
Relationships=pd.Series(Adults['relationship'])
RelCount=pd.Series(Relationships.groupby(Adults['relationship']).count())
print(RelCount)
print('*'*20)
print(RelCount.keys())
print('*'*20)
print(RelCount.get(key=' Other-relative'))




```
